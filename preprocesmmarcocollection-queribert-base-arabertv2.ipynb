```python
# https://huggingface.co/aubmindlab/araelectra-base-discriminator
# A preprocessing is recommended by the authors of AraELECTRA and AraBERT before training or testing on any dataset. 
!pip install arabert -q
from arabert.preprocess import ArabertPreprocessor
model_name="bert-base-arabertv2"
arabert_prep = ArabertPreprocessor(model_name=model_name)
#text = "و لن نبالغ إذا قلنا إن الهاتف أ و كمبيوتر  المكتب في زمننا هذا ضروري"
#arabert_prep.preprocess(text)
def preprocess(batch):
    for i in range(len(batch["text"])):
        batch["text"][i] = arabert_prep.preprocess(batch["text"][i])
    return batch
```

```python
from datasets import load_dataset
datasetQ = load_dataset('unicamp-dl/mmarco', 'queries-arabic', trust_remote_code=True)
datasetQ['train'][1]
```

```python
datasetQ= datasetQ.map(preprocess, batched=True)
```

```python
datasetC = load_dataset('unicamp-dl/mmarco', 'collection-arabic', trust_remote_code=True)
datasetC['collection'][100]
```

```python
datasetC= datasetC.map(preprocess, batched=True, num_proc=8)
```
